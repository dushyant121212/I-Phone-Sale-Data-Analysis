```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]
figure = px.bar(highest_rated,x=label,y=counts)
figure.update_layout()
figure.show()
#iphones = highest_rated['Product Name'].value_counts()
#label = iphones.index
#counts = highest_rated["Number Of Ratings"] / 1000   # divide values by 1000

#figure = px.bar(
 #   x=label,
  #  y=counts,
   # labels={'x': 'iPhone Model', 'y': 'Number Of Ratings (in Thousands)'}
#)

#figure.update_layout(xaxis_tickangle=-45)  # rotate x labels
#figure.show()
```

Which Iphone has the highest no. of reviews on flipkart

```python
df.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated["Number Of Reviews"]
figure = px.bar(highest_rated,x=label,y = counts,title= "Number of ratings of highest rated iPhones")
figure.show()
```

What is the relationship between the sale price of iphones and the number of ratings on Flipkart

```python
figure= px.scatter(data_frame=df,x="Number Of Ratings",y="Sale Price",size="Discount Percentage",trendline= 'ols',title="relationship between Sale price and no. of rating of Iphone")
figure.show()
```

What is the relationship between the discount percentage and the number of ratings of iphone on Flipkart 

```python
figure = px.scatter(data_frame=df,x="Number Of Ratings",y="Discount Percentage",size="Sale Price",trendline='ols',title="relationship between discount and Ratings of Iphone")
figure.show()
```

Figure out the least experience and most expensive iphones in the Indian market,along with all their specifications

```python
df.head()
```

```python
most_expensive = df.loc[df['Sale Price'].idxmax()]
least_expensive = df.loc[df['Sale Price'].idxmin()]

print("Most Expensive Product: ")
print(most_expensive)

print("Least Expensive Product: ")
print(least_expensive)
```
